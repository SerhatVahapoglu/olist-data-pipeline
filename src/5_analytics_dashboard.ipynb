```python
%sql
SELECT 
    c.customer_city,
    count(o.order_id) as total_orders
FROM end_to_end_project.olist_silver.orders o
JOIN end_to_end_project.olist_silver.customers c 
    ON o.customer_id = c.customer_id
GROUP BY c.customer_city
ORDER BY total_orders DESC
LIMIT 10
```

```python
%sql
SELECT 
    date_format(order_purchase_timestamp, 'yyyy-MM') as month_year,
    count(order_id) as total_orders
FROM end_to_end_project.olist_silver.orders
GROUP BY month_year
ORDER BY month_year
```
