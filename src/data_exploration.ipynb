```python
%sql 
SELECT
o.order_id,
i.price,
p.product_category_name
FROM end_to_end_project.olist_bronze.orders o
JOIN end_to_end_project.olist_bronze.order_items i ON o.order_id = i.order_id
JOIN end_to_end_project.olist_bronze.products p ON i.product_id = p.product_id
LIMIT 10;

```
